```python
import pandas as pd
import os
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('spam.csv', encoding='latin-1')
df
```

```python
df=df.drop(['Unnamed: 2','Unnamed: 3','Unnamed: 4'],axis='columns')
```

```python
df
```

```python
df['spam']=df['v1'].apply(lambda x: 1 if x=='spam' else 0)
df
```

```python
x_train,x_test,y_train,y_test=train_test_split(df.v2,df.spam,test_size=0.2)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vector=CountVectorizer()
```

```python
x_train_v=vector.fit_transform(x_train.values)
x_train_v
```

```python
x_train_v.shape
```

```python
dir(vector)
```

```python
vector.vocabulary_
```

```python
x_train_np=x_train_v.toarray()
x_train_np
```

```python
import numpy as np
np.where(x_train_np[0]!=0)
```

```python
x_train[0]
```

```python
from sklearn.naive_bayes import MultinomialNB
model=MultinomialNB()
model.fit(x_train_v,y_train)
```

```python
x_test_v=vector.transform(x_test)
x_test_v
```

```python
y_pred=model.predict(x_test_v)
y_pred
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
sms=[
    'hey can we get together to watch footbal game tomorrow',
    'Upto 20% discount on parking, exclusive offer just for you. Dont miss this reward'
]
```

```python
sms_numerical=vector.transform(sms)
predictions=model.predict(sms_numerical)
```

```python
for prediction in predictions:
    if prediction==1:
        print('spam')
    else:
        print('ham')
```
