```python
import pandas as pd
from sklearn.model_selection import train_test_split
import string
from string import digits
import re
from sklearn.utils import shuffle
import tensorflow as tf
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.layers import LSTM, Input, Dense,Embedding
from tensorflow.keras.models import Model,load_model
from tensorflow.keras.utils import plot_model
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.models import model_from_json
import pickle as pkl
import numpy as np
```

```python
# Lowercase
df['English'] = df['English'].str.lower()
df['Arabic'] = df['Arabic'].str.lower()

# Remove punctuation
punct_table = str.maketrans('', '', string.punctuation)
df['English'] = df['English'].apply(lambda x: ' '.join(w.translate(punct_table) for w in x.split()))
df['Arabic'] = df['Arabic'].apply(lambda x: ' '.join(w.translate(punct_table) for w in x.split()))

# Remove digits
digits = str.maketrans('', '', string.digits)
df['English'] = df['English'].apply(lambda x: ' '.join(w.translate(digits) for w in x.split()))
df['Arabic'] = df['Arabic'].apply(lambda x: re.sub(r'[\u0660-\u0669٠-٩]', '', x))

# Strip spaces
df['English'] = df['English'].str.strip()
df['Arabic'] = df['Arabic'].str.strip()
```

```python
# Add start and end tokens
df['Arabic'] = df['Arabic'].apply(lambda x: f"start {x} end")
```

```python
# Split data
X_train, X_test, y_train, y_test = train_test_split(df['English'], df['Arabic'], test_size=0.2)
```

```python
# Tokenization
input_tokenizer = Tokenizer()
input_tokenizer.fit_on_texts(X_train)
X_train_seq = input_tokenizer.texts_to_sequences(X_train)
X_test_seq = input_tokenizer.texts_to_sequences(X_test)

output_tokenizer = Tokenizer()
output_tokenizer.fit_on_texts(y_train)
y_train_seq = output_tokenizer.texts_to_sequences(y_train)
y_test_seq = output_tokenizer.texts_to_sequences(y_test)
```

```python
# Padding
max_input_len = max(len(seq) for seq in X_train_seq)
max_output_len = max(len(seq) for seq in y_train_seq)

X_train_pad = pad_sequences(X_train_seq, maxlen=max_input_len, padding='post')
X_test_pad = pad_sequences(X_test_seq, maxlen=max_input_len, padding='post')
y_train_pad = pad_sequences(y_train_seq, maxlen=max_output_len, padding='post')
y_test_pad = pad_sequences(y_test_seq, maxlen=max_output_len, padding='post')
```

```python
# One-hot encode y_train
vocab_target_size = len(output_tokenizer.word_index) + 1
y_train_onehot = tf.keras.utils.to_categorical(y_train_pad, num_classes=vocab_target_size)
y_test_onehot = tf.keras.utils.to_categorical(y_test_pad, num_classes=vocab_target_size)
```

```python
# Prepare decoder input (target sequence shifted by 1)
y_train_decoder_input = np.zeros_like(y_train_pad)
y_train_decoder_input[:, 1:] = y_train_pad[:, :-1]
```

```python
latent_dim = 50

encoder_inputs = Input(shape=(None,), name="encoder_inputs")
encoder_embedding = Embedding(input_dim=len(input_tokenizer.word_index)+1, output_dim=latent_dim, mask_zero=True)(encoder_inputs)
encoder_lstm, state_h, state_c = LSTM(latent_dim, return_state=True)(encoder_embedding)
encoder_states = [state_h, state_c]

decoder_inputs = Input(shape=(None,), name="decoder_inputs")
decoder_embedding = Embedding(input_dim=vocab_target_size, output_dim=latent_dim, mask_zero=True)(decoder_inputs)
decoder_lstm, _, _ = LSTM(latent_dim, return_sequences=True, return_state=True)(decoder_embedding, initial_state=encoder_states)
decoder_dense = Dense(vocab_target_size, activation='softmax')
decoder_outputs = decoder_dense(decoder_lstm)

model = Model([encoder_inputs, decoder_inputs], decoder_outputs)
model.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
# Prepare decoder input (target sequence shifted by 1)
y_train_decoder_input = np.zeros_like(y_train_pad)
y_train_decoder_input[:, 1:] = y_train_pad[:, :-1]
```

```python
# Train model
model.fit(
    [X_train_pad, y_train_decoder_input],
    y_train_onehot,
    batch_size=128,
    epochs=20,
    validation_split=0.2
)
```

```python
import random

def decode_seq(input_seq):
    state_values = encoder_model.predict(input_seq)
    target_seq = np.zeros((1, 1))
    target_seq[0, 0] = output_tokenizer.word_index['start']
    stop_condition = False
    decoded_sentence = []

    vocab_size = len(output_tokenizer.word_index) + 1  

    while not stop_condition:
        output_tokens, h, c = decoder_model.predict([target_seq] + state_values)
        sampled_token_index = np.argmax(output_tokens[0, -1, :])
        sampled_word = reverse_word_map_target.get(sampled_token_index, '')


        if sampled_word == '' or sampled_word == 'end' or len(decoded_sentence) > 70:
            if len(decoded_sentence) == 0:
                sampled_token_index = random.randint(1, vocab_size-1)
                sampled_word = reverse_word_map_target.get(sampled_token_index, 'junk')
                decoded_sentence.append(sampled_word)
                target_seq[0, 0] = sampled_token_index
                state_values = [h, c]
            else:
                stop_condition = True
        else:
            decoded_sentence.append(sampled_word)
            target_seq[0, 0] = sampled_token_index
            state_values = [h, c]

    return ' '.join(decoded_sentence).strip()


```
